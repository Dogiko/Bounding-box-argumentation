```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from bounding_box_argumentation import random_crop_square

def show_cv(img):
    plt.imshow(img[:,:,::-1])
    plt.show() #wrap
```

```python
box_start = np.array([50, 30])
box_end = np.array([230, 150])

img = cv2.imread("sample_cow.jpeg")
img = cv2.rectangle(img, tuple(box_start), tuple(box_end), (0,255,0), 2)
img = cv2.circle(img,tuple(((box_start + box_end)/2).astype(np.int)), 2, (255, 0, 0), -1)
show_cv(img)
```

```python
corp = random_crop_square(img, start=box_start, end=box_end, n_crop=1, resize=128, flip=True)

i = 0
show_cv(corp[i][0])
print(corp[i][1:])
```

```python
corp = random_crop_square(img, start=box_start, end=box_end, n_crop=10, resize=128, flip=True)

fig=plt.figure(figsize=(16, 16))
i=1
for crop_result in corp:
    fig.add_subplot(1, 10, i)
    i+=1
    plt.imshow(crop_result[0][:,:,::-1])
```
